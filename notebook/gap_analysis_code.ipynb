```python
import pandas as pd
import numpy as np

# Step 1: Parse Competitor Data
competitor_data = {
    'business': ['Root & Press – Café & Books', 'Acoustic Java Roastery & Tasting Room', 
                 'InHouse Coffee', 'BirchTree Bread Company', 'Nu Kitchen Worcester', 
                 'Coffee-Shop-Café', 'Blue Shades', 'Baker Shoe Co.', 'Dunkin', 'Starbucks'],
    'address': ['156 Shrewsbury St', '3 Brussels St', '225 Shrewsbury St', 
                '138 Green St', '335 Chandler St', '12 Thames St', '256 Park Ave', 
                '122 Warren Rd', '340 Grove St', '16 New Lancaster Rd'],
    'rating': [4.8, 4.6, 4.3, 4.6, 4.5, 4.0, 4.6, 4.4, 3.8, 3.7],
    'reviews': [593, 277, 721, 1600, 1900, 120, 145, 80, 260, 200],
    'zone': ['Downtown', 'Downtown', 'Downtown', 'North-East', 'Suburban', 
             'Downtown', 'Downtown', 'North-East', 'East', 'East'],
    'type': ['independent', 'independent', 'independent', 'independent', 'independent', 
             'independent', 'closed', 'independent', 'chain', 'chain']
}

competitors_df = pd.DataFrame(competitor_data)

# Step 2: Extract Market Fundamentals
market_fundamentals = {
    'population': 207000,
    'median_income': 71400,
    'population_density': 5527,
    'foot_traffic': {
        'downtown': 4000,
        'canal': 2500,
        'suburban': 1500,
        'northeast': 2000,
        'southwest': 1800,
        'east': 2200
    },
    'rent': {
        'downtown': 18,
        'canal': 15,
        'suburban': 12,
        'northeast': 14,
        'southwest': 13,
        'east': 12
    },
    'infrastructure_score': 7,  # out of 10
    'chain_dominance': 0.3  # 30% of businesses are chains
}

# Step 3: Calculate Zone Metrics
zones = {
    'Downtown': {
        'competitors': competitors_df[competitors_df['zone'] == 'Downtown'].shape[0],
        'avg_rating': competitors_df[competitors_df['zone'] == 'Downtown']['rating'].mean(),
        'total_reviews': competitors_df[competitors_df['zone'] == 'Downtown']['reviews'].sum(),
        'chain_count': competitors_df[(competitors_df['zone'] == 'Downtown') & (competitors_df['type'] == 'chain')].shape[0],
        'high_performer_count': competitors_df[(competitors_df['zone'] == 'Downtown') & (competitors_df['rating'] >= 4.5)].shape[0],
        'foot_traffic': market_fundamentals['foot_traffic']['downtown'],
        'rent': market_fundamentals['rent']['downtown'],
        'area_sq_mi': 1.5  # Approximate area
    },
    'Canal District': {
        'competitors': competitors_df[competitors_df['zone'] == 'Canal District'].shape[0],
        'avg_rating': competitors_df[competitors_df['zone'] == 'Canal District']['rating'].mean(),
        'total_reviews': competitors_df[competitors_df['zone'] == 'Canal District']['reviews'].sum(),
        'chain_count': competitors_df[(competitors_df['zone'] == 'Canal District') & (competitors_df['type'] == 'chain')].shape[0],
        'high_performer_count': competitors_df[(competitors_df['zone'] == 'Canal District') & (competitors_df['rating'] >= 4.5)].shape[0],
        'foot_traffic': market_fundamentals['foot_traffic']['canal'],
        'rent': market_fundamentals['rent']['canal'],
        'area_sq_mi': 2.0 
    },
    'Suburban': {
        'competitors': competitors_df[competitors_df['zone'] == 'Suburban'].shape[0],
        'avg_rating': competitors_df[competitors_df['zone'] == 'Suburban']['rating'].mean(),
        'total_reviews': competitors_df[competitors_df['zone'] == 'Suburban']['reviews'].sum(),
        'chain_count': competitors_df[(competitors_df['zone'] == 'Suburban') & (competitors_df['type'] == 'chain')].shape[0],
        'high_performer_count': competitors_df[(competitors_df['zone'] == 'Suburban') & (competitors_df['rating'] >= 4.5)].shape[0],
        'foot_traffic': market_fundamentals['foot_traffic']['suburban'],
        'rent': market_fundamentals['rent']['suburban'],
        'area_sq_mi': 3.0 
    },
    'North-East': {
        'competitors': competitors_df[competitors_df['zone'] == 'North-East'].shape[0],
        'avg_rating': competitors_df[competitors_df['zone'] == 'North-East']['rating'].mean(),
        'total_reviews': competitors_df[competitors_df['zone'] == 'North-East']['reviews'].sum(),
        'chain_count': competitors_df[(competitors_df['zone'] == 'North-East') & (competitors_df['type'] == 'chain')].shape[0],
        'high_performer_count': competitors_df[(competitors_df['zone'] == 'North-East') & (competitors_df['rating'] >= 4.5)].shape[0],
        'foot_traffic': market_fundamentals['foot_traffic']['northeast'],
        'rent': market_fundamentals['rent']['northeast'],
        'area_sq_mi': 2.5 
    },
    'South-West': {
        'competitors': competitors_df[competitors_df['zone'] == 'South-West'].shape[0],
        'avg_rating': competitors_df[competitors_df['zone'] == 'South-West']['rating'].mean(),
        'total_reviews': competitors_df[competitors_df['zone'] == 'South-West']['reviews'].sum(),
        'chain_count': competitors_df[(competitors_df['zone'] == 'South-West') & (competitors_df['type'] == 'chain')].shape[0],
        'high_performer_count': competitors_df[(competitors_df['zone'] == 'South-West') & (competitors_df['rating'] >= 4.5)].shape[0],
        'foot_traffic': market_fundamentals['foot_traffic']['southwest'],
        'rent': market_fundamentals['rent']['southwest'],
        'area_sq_mi': 2.8 
    },
    'East': {
        'competitors': competitors_df[competitors_df['zone'] == 'East'].shape[0],
        'avg_rating': competitors_df[competitors_df['zone'] == 'East']['rating'].mean(),
        'total_reviews': competitors_df[competitors_df['zone'] == 'East']['reviews'].sum(),
        'chain_count': competitors_df[(competitors_df['zone'] == 'East') & (competitors_df['type'] == 'chain')].shape[0],
        'high_performer_count': competitors_df[(competitors_df['zone'] == 'East') & (competitors_df['rating'] >= 4.5)].shape[0],
        'foot_traffic': market_fundamentals['foot_traffic']['east'],
        'rent': market_fundamentals['rent']['east'],
        'area_sq_mi': 2.2 
    }
}

# Calculate metrics for each zone
zone_metrics = []
for zone_name, data in zones.items():
    competitor_density = data['competitors'] / data['area_sq_mi']
    competition_quality_score = (data['avg_rating'] * 0.7) + (data['high_performer_count'] * 0.3)
    chain_dominance_ratio = data['chain_count'] / data['competitors'] if data['competitors'] > 0 else 0
    demand_signal = (data['foot_traffic'] * data['avg_rating']) / 1000
    
    # Market Saturation Index: (Competitors × Quality) / Demand
    saturation_index = (data['competitors'] * competition_quality_score) / demand_signal if demand_signal > 0 else 0
    
    # Viability Score: weighted based on key factors
    viability_score = (
        (1 - saturation_index) * 0.3 +  # Lower saturation better
        (demand_signal / 1000) * 0.2 +   # Higher demand better
        (1 - chain_dominance_ratio) * 0.15 +  # Lower chain dominance better
        (data['avg_rating'] / 5) * 0.15 +  # Higher ratings better
        ((market_fundamentals['infrastructure_score'] / 10) * 0.10) +  # Infrastructure quality
        (1 - (data['rent'] / market_fundamentals['rent']['downtown'])) * 0.10  # Lower rent better
    )
    
    zone_metrics.append({
        'zone': zone_name,
        'competitor_count': data['competitors'],
        'competitor_density': round(competitor_density, 2),
        'avg_rating': round(data['avg_rating'], 2),
        'total_reviews': data['total_reviews'],
        'competition_quality_score': round(competition_quality_score, 2),
        'chain_dominance_ratio': round(chain_dominance_ratio, 2),
        'high_performer_count': data['high_performer_count'],
        'foot_traffic_per_day': data['foot_traffic'],
        'rent_per_sf': data['rent'],
        'demand_signal': round(demand_signal, 2),
        'saturation_index': round(saturation_index, 2),
        'viability_score': round(viability_score, 2)
    })

# Convert to DataFrame
zones_df = pd.DataFrame(zone_metrics)

# Step 4: Zone Categorization
def categorize_zone(row):
    if row['saturation_index'] > 1.5:
        return 'SATURATED'
    elif row['saturation_index'] > 0.8:
        return 'MODERATE'
    else:
        return 'OPPORTUNITY'

def assess_risk(row):
    if row['rent_per_sf'] > 16:
        return 'HIGH'
    elif row['rent_per_sf'] > 13:
        return 'MEDIUM'
    else:
        return 'LOW'

def assess_investment_tier(row):
    if row['saturation_index'] < 0.8 and row['viability_score'] > 0.7:
        return 'HIGH'
    elif row['saturation_index'] < 1.0 and row['viability_score'] > 0.6:
        return 'MEDIUM'
    else:
        return 'LOW'

def identify_customer_segment(row):
    if 'Suburban' in row['zone'] or 'North-East' in row['zone']:
        return 'Professionals / Families'
    elif 'Canal' in row['zone']:
        return 'Students'
    elif 'Downtown' in row['zone']:
        return 'Young Professionals'
    else:
        return 'General'

zones_df['category'] = zones_df.apply(categorize_zone, axis=1)
zones_df['risk_level'] = zones_df.apply(assess_risk, axis=1)
zones_df['investment_tier'] = zones_df.apply(assess_investment_tier, axis=1)
zones_df['best_customer_segment'] = zones_df.apply(identify_customer_segment, axis=1)

# Step 5: Rank Top Zones
zones_df['rank'] = zones_df['viability_score'].rank(ascending=False)

# Step 6: Output Tables
print("=== ZONE METRICS ===")
print(zones_df[['zone', 'competitor_count', 'competitor_density', 'avg_rating', 
               'chain_dominance_ratio', 'saturation_index', 'viability_score', 'category']].round(2))

print("\n=== TOP 3 RECOMMENDED ZONES ===")
top_zones = zones_df.sort_values('viability_score', ascending=False).head(3)
print(top_zones[['zone', 'viability_score', 'saturation_index', 'competitor_count', 'category', 'risk_level']])

print("\n=== RISK ASSESSMENT MATRIX ===")
risk_matrix = zones_df[['zone', 'rent_per_sf', 'saturation_index', 'risk_level']]
print(risk_matrix)

print("\n=== STRATEGIC RECOMMENDATIONS ===")
print("1. The most viable location is:", top_zones.iloc[0]['zone'])
print("   - Viability Score:", top_zones.iloc[0]['viability_score'])
print("   - Saturation Index:", top_zones.iloc[0]['saturation_index'])
print("   - Risk Level:", top_zones.iloc[0]['risk_level'])
print("   - Best Customer Segment:", top_zones.iloc[0]['best_customer_segment'])
print("\n2. Second best opportunity is:", top_zones.iloc[1]['zone'])
print("   - Viability Score:", top_zones.iloc[1]['viability_score'])
print("   - Saturation Index:", top_zones.iloc[1]['saturation_index'])
print("   - Risk Level:", top_zones.iloc[1]['risk_level'])
print("   - Best Customer Segment:", top_zones.iloc[1]['best_customer_segment'])
print("\n3. Third best opportunity is:", top_zones.iloc[2]['zone'])
print("   - Viability Score:", top_zones.iloc[2]['viability_score'])
print("   - Saturation Index:", top_zones.iloc[2]['saturation_index'])
print("   - Risk Level:", top_zones.iloc[2]['risk_level'])
print("   - Best Customer Segment:", top_zones.iloc[2]['best_customer_segment'])
```
